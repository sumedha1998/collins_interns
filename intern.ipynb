# Read data

```python
import pandas as pd
import numpy as np
data=pd.read_csv("sales.csv")
data.head(5)
```

```python
print(data.size)
data.shape
```

```python
data.info()
```

# one hot encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df=data

df['Opportunity Result']= le.fit_transform(df["Opportunity Result"])
df['Supplies Subgroup']= le.fit_transform(df["Supplies Subgroup"])
df['Supplies Group']= le.fit_transform(df["Supplies Group"])
df['Region']= le.fit_transform(df["Region"])
df['Route To Market']=le.fit_transform(df["Route To Market"])
df['Competitor Type']=le.fit_transform(df["Competitor Type"])
```

```python
df.head()
```

# changing data type( float64 to int64)

```python
df["Ratio Days Identified To Total Days"]=df["Ratio Days Identified To Total Days"].astype(np.int64)
df["Ratio Days Validated To Total Days"]=df["Ratio Days Validated To Total Days"].astype(np.int64)
df["Ratio Days Qualified To Total Days"]=df["Ratio Days Qualified To Total Days"].astype(np.int64)
```

# checking for null values

```python
print(df.isna())
```

```python
from sklearn.preprocessing import MinMaxScaler

mms=MinMaxScaler()
df1=mms.fit_transform(df)

print(df1)
df1=pd.DataFrame(df1)
print(df1.head())
```

# Regresion Model

```python
from sklearn.model_selection import train_test_split

Y=df1[18]
X=df1.drop([18],axis=1)

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
print("Y_pred")
print(Y_pred)
print()
print("Y_test")
print(Y_test)
```
